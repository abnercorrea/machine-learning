```python
import numpy as np
import scipy as sp
import pandas as pd

import sklearn as skl
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Normalizer

import tensorflow as tf
import tensorflow.experimental.numpy as tnp
from  tensorflow.python.ops.numpy_ops import np_config 
import tensorboard
```

```python
from abnercorrea.tensorflow.linear.ridge import RidgeRegression
from abnercorrea.tensorflow.linear.linear_regression import LinearRegression
from abnercorrea.tensorflow.neighbors import KNearestNeighborsClassifierTF

from abnercorrea.numpy.util.data_vis import plot_alpha_scores
```

```python
np_config.enable_numpy_behavior()
```

```python
iris = sklearn.datasets.load_iris()
```

```python
X, y = iris.data, iris.target
```

```python
xtr, xte, ytr, yte = train_test_split(X, y, test_size=0.2)
```

```python
norm = Normalizer()
xtrn, xten = norm.transform(xtr), norm.transform(xte)
```

```python
knn = KNearestNeighborsClassifierTF()
knn_norm = KNearestNeighborsClassifierTF()
```

```python
alpha_grid = np.arange(1, 101)
```

```python
%time knn.fit(xtr, ytr, alpha_grid)
```

```python
plot_alpha_scores(alpha_grid, knn.k_scores)
```

```python
knn.k
```

```python
knn.score(xte, yte)
```

```python
%time knn_norm.fit(xtrn, ytr, alpha_grid)
```

```python
plot_alpha_scores(alpha_grid, knn_norm.k_scores)
```

```python
knn_norm.k
```

```python
knn_norm.k = 1
```

```python
knn_norm.score(xten, yte)
```
